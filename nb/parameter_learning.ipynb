# Parameter learning


In order to model and simulate the hydrology over the Alpine region with Wflow, the spatial domain is discretized by a regular grid, where each cell is associated with a vector of Wflow's model parameters  $ W_\theta $.

Several biogeophysical attributes $ \Phi $ are used to estime $ W_\theta $, most of the time by means of pedotransfer functions.

$ \Phi $ are most of the time defined over a regular grid as well.

The spatial resolution of $ W_\theta $ is usually $ \geq $ than $ \Phi $
 
Thefore, each cell of $ W_\theta $ can spatially overlay a greater number of higher resolution $ \Phi $ cells.

1) Stack the $ \Phi $ layers (decide whether to resample to highest resolution). Model grid cells are referenced by an index, available either in 2D matrix format or 1D vector. The same indices should be associated to the $ \Phi $ cells, therefore we need an higher resolution 2D matrix with the same indices of the Wflow matrix as defined by a spatial join.

2) The structure of the AI layers: anything that can encode the information of many layers of $ \Phi $ to smaller sized latent vector $ \mathbf{z} $ with an inductive bias for spatial patters (CNNs)

VAEs seem interesting. They allow for production of distribution moments (i.e., mean and variance in case of a normal distribution).
